```python
#Q1: To check if there are null values and drop them
print(df.isnull().sum())
df.dropna() #drop any row with NaN
df.dropna(axis=1) #drop any column with NaN
```

```python
#Q2: To find unique instances of weather types
weather=df['Weather'].value_counts()
dfweather=pd.DataFrame(weather)
dfweather = dfweather.reset_index()
dfweather.columns = ['Weather', 'Frequency'] # change column names
dfweather
```

```python
# prompt: Using dataframe dfweather: Frequency

# Calculate the frequency of each weather condition
weather_frequencies = dfweather['Weather'].value_counts()

# Display the frequencies
weather_frequencies
```

```python
#Q3: To rename column named 'Weather' to Weather Condition'
df.rename(columns = {'Weather' : 'Weather_Condition'}, inplace=False)
```


## Step 5: Data Visualization  
We will generate plots to better understand the weather trends.  


```python
#Q6: To find the variance of pressure
print(df['Press_kPa'].var())
```

```python
#Q7: To find the days on which wind speed was less than or equal to 30 km/hr and temperature was greater than 0 C
windspeed=df[(df['Wind Speed_km/h'] <=30) & (df['Temp_C']>0)]
print(windspeed['Date/Time'])
print(windspeed.shape)
```

```python
#Q8: To find the date and temperatures for all instances when snow was recorded
snow=df[df['Weather_Condition'] == 'Snow']
snow.loc[:, ['Date/Time', 'Temp_C']]
```
